```python
import numpy as np
from ml_models import models_dict
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
plt.style.use("ggplot")
```

```python
def data_preprocessor(fname, modify=None):
    with open(fname, 'r') as f:
        rules = f.read().split('\n')

    X,  Y = [], []
    for rule in rules:
        (UA, OA, P) = rule.split(';')
        UA = list(map(int, UA.split(',')))
        OA = list(map(int, OA.split(',')))
        if modify is None:
            X.append(UA + OA)
        elif modify == "extend":
            extn = [0, 0, 0]
            extn[0] = int(UA[1] == OA[1])
            extn[1] = int(UA[2] == OA[2])
            extn[2] = int(UA[3] == OA[3])
            X.append(UA + OA + extn)
        elif modify == "compress":
            extn = [0, 0, 0]
            extn[0] = int(UA[1] == OA[1])
            extn[1] = int(UA[2] == OA[2])
            extn[2] = int(UA[3] == OA[3])
            X.append([UA[0], OA[0]] + extn)
        Y.append(int(P))
    return (np.array(X), np.array(Y))
```

```python
Xtrain, Ytrain = data_preprocessor("abac-cat-corrected-v1.txt")
Xtest, Ytest = data_preprocessor("test-v1.txt")

l = Xtrain.shape[0]
X = np.concatenate((Xtrain, Xtest), axis=0)
X_scaled = TSNE(n_components=2, n_jobs=-1).fit_transform(X)

Xtrain, Xtest = X_scaled[:l], X_scaled[l:]
Xtrain_pos, Xtrain_neg = Xtrain[Ytrain==1], Xtrain[Ytrain==0]
Xtest_pos, Xtest_neg = Xtest[Ytest==1], Xtest[Ytest==0]
```

```python
plt.scatter(Xtrain_pos[:,0], Xtrain_pos[:,1], color='g', marker='*', label="train +")
plt.scatter(Xtrain_neg[:,0], Xtrain_neg[:,1], color='r', marker='*', label="train -")
plt.scatter(Xtest_pos[:,0], Xtest_pos[:,1], color='y', marker='*', label="test +")
plt.scatter(Xtest_neg[:,0], Xtest_neg[:,1], color='m', marker='*', label="test -")
plt.title("data tsne plot")
plt.legend(loc="best")
plt.show()
```
